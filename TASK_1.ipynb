```python
unique_cnt=df['job'].unique()
count_list=[]
for i in unique_cnt:
    con=df['job']==i
    val=len(df[con])
    count_list.append(val)
```

```python
job_df=pd.DataFrame(zip(unique_cnt,count_list),columns=['JOB','COUNT'])
job_df
```

```python
unique_cnt=df['education'].unique()
count_list=[]
for i in unique_cnt:
    con=df['education']==i
    val=len(df[con])
    count_list.append(val)
```

```python
education_df=pd.DataFrame(zip(unique_cnt,count_list),columns=['EDUCATION','COUNT1'])
education_df
```

```python
unique_cnt=df['contact'].unique()
count_list=[]
for i in unique_cnt:
    con=df['contact']==i
    val=len(df[con])
    count_list.append(val)
```

```python
contact_df=pd.DataFrame(zip(unique_cnt,count_list),columns=['CONTACT TYPE','USERS'])
contact_df
```

```python
unique_cnt=df['marital'].unique()
count_list=[]
for i in unique_cnt:
    con=df['marital']==i
    val=len(df[con])
    count_list.append(val)
```

```python
marital_df=pd.DataFrame(zip(unique_cnt,count_list),columns=['MARITAL STATUS','PEOPLE'])
marital_df
```

```python
unique_cnt=df['poutcome'].unique()
count_list=[]
for i in unique_cnt:
    con=df['poutcome']==i
    val=len(df[con])
    count_list.append(val)
    
poutcome_df=pd.DataFrame(zip(unique_cnt,count_list),columns=['POUTCOME','STATUS'])
poutcome_df
```

**BAR CHART**

- Here We plot BarChart with two columns :
    
    - JOB
    
    - COUNT
    
- It Will give visualization of data distributed over different jobs

```python
plt.figure(figsize=(16,6))
plt.bar('JOB','COUNT',data=job_df,label="JOB")
plt.title("BAR CHART - COUNT OF DIFFERENT CATEGORY JOBS")
plt.xlabel("JOB NAMES")
plt.ylabel("COUNT OF PEOPLE")
plt.legend(['JOB'])
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.bar('EDUCATION','COUNT1',data=education_df)
plt.title("BAR CHART - EDUCATION")
plt.xlabel("EDUCATION")
plt.ylabel("COUNT OF PEOPLE")
plt.legend(['EDUCATION'])
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.bar('CONTACT TYPE','USERS',data=contact_df)
plt.title("BAR CHART - CONTACT")
plt.xlabel("CONTACT TYPE")
plt.ylabel("USERS")
plt.legend(['CONTACT TYPE'])
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.bar('MARITAL STATUS','PEOPLE',data=marital_df)
plt.title("BAR CHART - MARITAL STATUS")
plt.xlabel("MARITAL STATUS")
plt.ylabel("PEOPLE")
plt.legend(['MARITAL STATUS'])
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.bar('POUTCOME','STATUS',data=poutcome_df)
plt.title("BAR CHART - POUTCOME")
plt.xlabel("OUTCOME")
plt.ylabel("COUNT OF PEOPLE")
plt.legend(['POUTCOME'])
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.subplot(1,4,1)
plt.bar('JOB','COUNT',data=job_df)
plt.title("BAR CHART - COUNT OF DIFFERENT CATEGORY JOBS")
plt.xlabel("JOB NAMES")
plt.ylabel("COUNT OF PEOPLE")
plt.legend(['JOB'])
plt.show()

plt.figure(figsize=(16,6))
plt.subplot(1,4,2)
plt.bar('EDUCATION','COUNT1',data=education_df)
plt.title("BAR CHART - EDUCATION")
plt.xlabel("EDUCATION")
plt.ylabel("COUNT OF PEOPLE")
plt.legend(['EDUCATION'])
plt.show()

plt.figure(figsize=(16,6))
plt.subplot(1,4,3)
plt.bar('MARITAL STATUS','PEOPLE',data=marital_df)
plt.title("BAR CHART-MARITAL STATUS")
plt.xlabel("MARITAL STATUS ")
plt.ylabel("USERS")
plt.legend(['MARITAL STATUS'])
plt.show()

plt.figure(figsize=(16,6))
plt.subplot(1,4,4)
plt.bar('POUTCOME','STATUS',data=poutcome_df)
plt.title("BAR CHART - POUTCOME")
plt.xlabel("OUTCOME")
plt.ylabel("COUNT OF PEOPLE")
plt.legend(['POUTCOME'])
plt.show()

```
